# Problem statement is predicting the sales price of house

# We need to concat train and test then we need to give salesprice value as 99
# we preprocess all data thenn we seperate the test data which contain saleprce=99
# then we remove saleprice from test

```python
import pandas as pd
```

```python
train_data=pd.read_csv("D:\\Kaggle Projects\\House price\\train.csv")
train_data
```

```python
test_data=pd.read_csv("D:\\Kaggle Projects\\House price\\test.csv")
test_data
```

```python
99 in train_data['SalePrice'].values
```

```python
Sale1=[99]*1459 #creating SalePrice =99 for all testdata
test_data['SalePrice']=Sale1
```

```python
test_data.head(2)
```

```python
#Concatin train and test data axis=0 means upper side train data followed by test at down
data=pd.concat([train_data,test_data],axis=0)
data.shape
```

```python
data_cols=data.columns
data_cols
```

```python
#finding null values
null_val=[]
null_cols=[]
for i in range(81):
    print(data_cols[i]," : ",data[data_cols[i]].isnull().sum())
    null_val.append(data[data_cols[i]].isnull().sum()) #here we storing all null_values in list

for j in range(81):
    if null_val[j]>0 :
        null_cols.append(data_cols[j]) # here we storing column names which contains null values
```

```python
len(null_cols)  #We having total 19 columns which containing null values
```

```python
# Lets check which column has how many null values   
for i in range(len(null_cols)):
    print(null_cols[i]," : ",data[null_cols[i]].isnull().sum())
```

```python
#Now we seen that some columns contain more than 33% null data
#So we need to drop that columns from our dataset
#here totaldata is 1500 so 40% is 650 
deleting_columns=[]
for i in range(len(null_cols)):
    if data[null_cols[i]].isnull().sum()>650:
        deleting_columns.append(null_cols[i])   
deleting_columns        
```

```python
data.drop(deleting_columns,axis=1,inplace=True)
```

```python
data.shape #before we have 81 columns now we have 76 columns
```

```python
#Here we are storing column names in which we need to fill value & which are not deleted.
fill_null_cols=[]
fill_null_cols=null_cols
for dc in deleting_columns:
    fill_null_cols.remove(dc)
fill_null_cols
```

```python
#we can fill null continues data with median.
#null categorical data with mode.
#here in this examplefor deciding mean or median we consider if the count of unique() values -
#-in any column is greater than 50 then it belongs to continues data otherwise it is categorical.
```

```python
#Here len() for counting length of each unique array data member
for i in range(len(fill_null_cols)):
    print(fill_null_cols[i]," : ",len(data[fill_null_cols[i]].unique()))
    
    if len(data[fill_null_cols[i]].unique())>50:
        #continues data replacing here by median
        data[fill_null_cols[i]]=data[fill_null_cols[i]].fillna(data[fill_null_cols[i]].median())
    else:
        #categorical data replacing here by mode
        data[fill_null_cols[i]]=data[fill_null_cols[i]].fillna(data[fill_null_cols[i]].mode()[0])
```

```python
data.isnull().sum()
```

```python
# columns names which contains catagorical value
data_choosen=data.columns
cat_cols1=[]
for i in data_choosen:
    if data[i].dtypes=='object':
        cat_cols1.append(i)
cat_cols1
```

```python
len(cat_cols1)
```

```python
for i in cat_cols1:
    print(str(i)+"  "+str(len(data[i].unique())))
```

```python
data.shape
```

```python
#Applying One hot encoding on categorical data
from sklearn.preprocessing import OneHotEncoder
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
enc_df = pd.DataFrame(enc.fit_transform(data[cat_cols1]).toarray())
```

```python
data1=pd.concat([data,enc_df],axis=1)
```

```python
data1.shape
```

```python
#Now drop the Cat_cols
data1.drop(columns=cat_cols1,axis=1,inplace=True)
```

```python
data1
```

```python
data1['SalePrice']=data1['SalePrice'].astype('int64')
```

# Now again we split the Train and Test data where Testdata['SalePrice']==99
